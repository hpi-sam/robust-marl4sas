```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from ast import literal_eval
from sklearn.metrics import auc
import numpy as np
from matplotlib.ticker import MaxNLocator
```

```python
df1 = pd.read_csv('output/cumfix100_runs_highest_utility.csv').reset_index()
df2 = pd.read_csv('output/cumfix100_runs_highest_utility_fixingcanfail.csv').reset_index()

df3 = pd.read_csv('output/cumfix100_runs_highest_cost.csv').reset_index()
df4 = pd.read_csv('output/cumfix100_runs_highest_cost_fixingcanfail.csv').reset_index()

df5 = pd.read_csv('output/cumfix100_runs_highest_random.csv').reset_index()
df6 = pd.read_csv('output/cumfix100_runs_highest_random_fixingcanfail.csv').reset_index()
```

## Possible Fix Failures vs Perfect Fixing

```python
#Look away!
df1["cumsum"] = df1['cum_util'].cumsum()
df2["cumsum"] = df2['cum_util'].cumsum()
df3["cumsum"] = df3['cum_util'].cumsum()
df4["cumsum"] = df4['cum_util'].cumsum()
df5["cumsum"] = df5['cum_util'].cumsum()
df6["cumsum"] = df6['cum_util'].cumsum()
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='index', y='cumsum', data=df1, ax=ax, ci=None, drawstyle='steps', label='Always Fix')
sns.lineplot(x='index', y='cumsum', data=df2, ax=ax, ci=None, drawstyle='steps', label='Fixing can Fail')
ax.set(xlabel='Status', ylabel='Utility', title='Cumulative Utility - Fix-Priorisation: Utility' );
fig.savefig('output/cum_utility.png', dpi=300)
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='index', y='cumsum', data=df3, ax=ax, ci=None, drawstyle='steps', label='Always Fix')
sns.lineplot(x='index', y='cumsum', data=df4, ax=ax, ci=None, drawstyle='steps', label='Fixing can Fail')
ax.set(xlabel='Status', ylabel='Utility', title='Cumulative Utility - Fix-Priorisation: Cost' );
fig.savefig('output/cum_cost.png', dpi=300)
```

```python
fig, ax = plt.subplots()
sns.lineplot(x='index', y='cumsum', data=df5, ax=ax, ci=None, drawstyle='steps', label='Always Fix')
sns.lineplot(x='index', y='cumsum', data=df6, ax=ax, ci=None, drawstyle='steps', label='Fixing can Fail')
ax.set(xlabel='Status', ylabel='Utility', title='Cumulative Utility - Fix-Priorisation: Random' );
fig.savefig('output/cum_random.png', dpi=300)
```
